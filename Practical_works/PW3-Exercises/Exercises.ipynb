## Operator Overloading 

### Exercise 1: Implement a complex number class. Overload the addition and subtraction operators to add and subtract two complex numbers.

```python
class Complexnumber: 
    def __init__(self, real, imaginary):
        self.real = real
        self.imaginary = imaginary 

    def __add__(self, other):
        real = self.real + other.real
        imaginary = self.imaginary + other.imaginary
        return Complexnumber(real, imaginary)
    
    def __sub__(self, other):
        real = self.real - other.real
        imaginary = self.imaginary - other.imaginary 
        return Complexnumber(real, imaginary)
    
    def __str__(self):
        return f"{self.real} + {self.imaginary}i"
```

```python
# Define two complex numbers to check if the addition and subtraction work correctly
c1 = Complexnumber(3, 4)
c2 = Complexnumber(1, 2)

# Test addition
result_add = c1 + c2
print(f"The sum of your two complex numbers is: {result_add}")

# Test Subtraction
result_sub = c1 - c2
print(f"The difference of your two complex numbers is: {result_sub}")
```

What I learn from this exercise: 
- Class is an "Object" that is defined with operations called "Methods"; here we have addition __add__, subtraction __sub__, and display __str__.
- We can also defined numbers through the class and perform the defined operations. 
- No overthinking on self, its a convention that helps with the format later on in the code.

### Exercise 3: Create a class Vector that represents a vector in space, with x, y, and z as its attributes. Overload the * operator to return the dot product of two vectors, such that vector1 * vector2 returns the dot product of the two input vectors.

```python
class Vector: 
    def __init__(self, x, y, z):
        self.x = x
        self.y = y 
        self.z = z
    
    def __mul__(self, other):
        dot_product = self.x * other.x + self.y * other.y + self.z * other.z
        return dot_product
```

```python
vector1 = Vector(1, 2, 3)
vector2 = Vector(4, 5, 6)

scalar_product = vector1.__mul__(vector2) # equivalent to vector1 * vector2
print(f"The scalar product of the two vectors is: {scalar_product}")
```

What I learned: 
- When i define a sequence of operations I can define a name like dot_product that takes the information of the result of the operation.

### Exercise 4: Overload all comparison operators (<, >, ==, !=, >=, <=) for a Circle class based on radius.

```python
class Circle: 

    def __init__(self, radius):
        self.radius = radius 

    def __lt__(self, other): # less than operator
        return self.radius < other.radius 

    def __gt__(self, other): # greater than operator
        return self.radius > other.radius
    
    def __eq__(self, other): # equal to operator
        return self.radius == other.radius
    
    def __ne__(self, other): # not equal to operator
        return self.radius != other.radius
    
    def __le__(self, other): # less than or equal to operator
        return self.radius <= other.radius 
    
    def __ge__(self, other): # greater than or equal to operator    
        return self.radius >= other.radius      
```

```python
Circle1 = Circle(5)
Circle2 = Circle(10)
Circle3 = Circle(2)
```

```python
print (Circle1 < Circle2) # True
print (Circle1 > Circle3) # True
print (Circle1 == Circle3) # False
print (Circle1 != Circle3) # True
print (Circle1 <= Circle2) # True
print (Circle1 >= Circle3) # True
```

### Exercise 5: Implement a 2D point class. Overload the comparison operators to compare two points based on their Euclidean distance from the origin.

```python
class Point:

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def Euclidean_distance(self):
       return (self.x ** 2 + self.y ** 2) ** 0.5
```

```python
point1 = Point(3, 4)
point2 = Point(6, 8)
print(f"The Euclidean distance of point 1 from the origin is: {point1.Euclidean_distance()}")
print(f"The Euclidean distance of point 2 from the origin is: {point2.Euclidean_distance()}")  

if point1.Euclidean_distance() < point2.Euclidean_distance():
    print("Point 1 is closer to the origin than Point 2")
elif point1.Euclidean_distance() > point2.Euclidean_distance():
    print("Point 2 is closer to the origin than Point 1")
else:
    print("Both points are equidistant from the origin")
```

## Inheritance exercises

### Exercise 1: Create a class Vehicle that has a method drive that returns the string "Driving a vehicle". Create two subclasses, Car and Bicycle, that override this method to return "Driving a car" and "Riding a bicycle", respectively.

```python
class Vehicle: 

    def __init__(self):
        pass

    def __drive__(self):
        print("Driving the vehicle...")
```

```python
v1 = Vehicle()
v1.__drive__()
```

```python
# Now we create the subclasses: Car and Bicycle. They inherit from the Vehicle class and override the __drive__ method to return "Driving the car" and "Riding the bicycle" respectively. 

class Car(Vehicle):

    def __drive__(self):
        print("Driving the car...")

class Bicycle(Vehicle):

    def __drive__(self):
        print("Riding the bicycle...")
```

```python
c1 = Car()
c1.__drive__()

b1 = Bicycle()
b1.__drive__()
```

What I learn: 
- Inheritance says that Child classes inherit methods from Parent classes, and they can override the methods of parents classes. 

### Exercise 2: 
Create a class Person with attributes: name, age, address and method: introduce.

Create a subclass Student of the class Person that has an attribute: field_of_study and a personalized method: introduce.

Create another subclass Employee of the class Person that has an attribute: company and a personalized method: introduce.



```python
class Person:

    def __init__(self, name, age, address):
        self.name = name
        self.age = age
        self.address = address

    def __introduyce__(self):
        print(f"Hello, my name is {self.name}, I am {self.age} years old and I live at {self.address}.")
```

```python
person1 = Person("John", 30, "123 Main St")
person1.__introduyce__()
```

```python
class Student(Person):

    def __init__(self, name, age, address, field_of_study):
        super().__init__(name, age, address)
        self.field_of_study = field_of_study

    def __introduce__(self):
        print(f"Hello, my name is {self.name}, I am {self.age} years old and I live at {self.address}. I study {self.field_of_study}.")
```

```python
student1 = Student("John", 30, "123 Main St", "Computer Science")
student1.__introduce__()
```

```python
class Employee(Person):

    def __init__(self, name, age, address, job_title):
        super().__init__(name, age, address)
        self.job_title = job_title

    def __introduce__(self):
        print(f"Hello, my name is {self.name}, I am {self.age} years old and I live at {self.address}. I work as a {self.job_title}.")
```

```python
employee1 = Employee("Jean", 28, "Paris", "Software Engineer at Apple")
employee1.__introduce__()
```
